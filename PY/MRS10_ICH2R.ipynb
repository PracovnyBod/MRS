## ICH SS2R

```python
# Dva nezavisle poly
p_1 = -1
p_2 = -2

A = 2 / (p_1 - p_2)
B = 2 / (p_2 - p_1)

# Súradnice bodov na x-ovej osi
plotData_x = np.arange(0, 10, 0.1)

# Výpočet hodnôt na y-ovej osi v zmysle danej časovej funkcie
plotData_y = A * np.exp(p_1 * plotData_x) + B * np.exp(p_2 * plotData_x)

# Kreslenie grafu
plt.plot(plotData_x, plotData_y)
plt.xlabel('čas $t$')
plt.ylabel('$y(t)$')
plt.show()
```
