```python
# Dva nezavisle poly

# Súradnice bodov na x-ovej osi
plotData_x = np.arange(0, 10, 0.1)

# Výpočet hodnôt na y-ovej osi v zmysle danej časovej funkcie
plotData_y = 1 - 2 * np.exp(-1 * plotData_x) + 1 * np.exp(-2 * plotData_x)

# Kreslenie grafu
plt.plot(plotData_x, plotData_y)
plt.xlabel('čas $t$')
plt.ylabel('$y(t)$')
plt.show()

# Obrázok pre hlavný text
figName = 'PCH_SS2R_v1'
figNameNum = 0
exec(open('./figjobs/MRS10_figJob_01.py', encoding='utf-8').read())
```

```python
# dvojnasobny pol

# Súradnice bodov na x-ovej osi
plotData_x = np.arange(0, 10, 0.1)

# Výpočet hodnôt na y-ovej osi v zmysle danej časovej funkcie
plotData_y = 1 - plotData_x * np.exp(-1 * plotData_x) - 1 * np.exp(-1 * plotData_x)


# Kreslenie grafu
plt.plot(plotData_x, plotData_y)
plt.xlabel('čas $t$')
plt.ylabel('$y(t)$')
plt.show()

# Obrázok pre hlavný text
figName = 'PCH_SS2R_v2'
figNameNum = 0
exec(open('./figjobs/MRS10_figJob_01.py', encoding='utf-8').read())
```

```python
# komplexne zdruzene poly

# Súradnice bodov na x-ovej osi
plotData_x = np.arange(0, 10, 0.05)

# Výpočet hodnôt na y-ovej osi v zmysle danej časovej funkcie
plotData_y = 1 -  np.exp(-(3/2)*plotData_x) * ( np.cos((3/2)*np.sqrt(3)*plotData_x) + (np.sqrt(3)*np.sin((3/2)*np.sqrt(3)*plotData_x)/3 ))

# Kreslenie grafu
plt.plot(plotData_x, plotData_y)
plt.xlabel('čas $t$')
plt.ylabel('$y(t)$')
plt.show()

# Obrázok pre hlavný text
figName = 'PCH_SS2R_v3'
figNameNum = 0
exec(open('./figjobs/MRS10_figJob_01.py', encoding='utf-8').read())
```
