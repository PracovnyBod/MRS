## PCH SS1R

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametre
b_0 = 1
a_0 = 1

# Súradnice bodov na x-ovej osi
plotData_x = np.arange(0, 10, 0.1)

# Výpočet hodnôt na y-ovej osi v zmysle danej časovej funkcie
plotData_y = (b_0/a_0) * (1 - np.exp(-a_0 * plotData_x))
```

Kreslenie jednoduchého obrázku

```python
# Kreslenie grafu
plt.plot(plotData_x, plotData_y)
plt.xlabel('čas $t$')
plt.ylabel('$y(t)$')
plt.show()
```
