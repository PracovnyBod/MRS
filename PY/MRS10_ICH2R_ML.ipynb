```matlab
p_1 = -1;
p_2 = -2;

polyA = conv([1, -p_1], [1, -p_2]);

syms s
As = polyA(1)*s^2 + polyA(2)*s + polyA(3);
Bs = polyA(end);

Gs = Bs/As;

y = ilaplace(Gs)
```

```matlab
G = tf(polyA(end), polyA)
impulse(G)
```
